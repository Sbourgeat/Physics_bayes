# Bayesian analysis

```python
import bambi as bmb
import arviz as az
import bambi as bmb
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
az.style.use("arviz-white")
np.random.seed(1111)
```

```python
df_bayes = filtered_df[["en_ns", "epsilon_critical"]].dropna()
#df_bayes = df_merged[["epsilon_critical", "en_ns"]]
df_bayes = df_bayes.dropna()
#df_bayes["bias_left_handedness"] = np.abs(df_bayes["bias_left_handedness"])


def standardize(series):
    """Standardize a pandas series"""
    return (series - series.mean()) / series.std()

for i in df_bayes.columns:
    df_bayes[i] = standardize(df_bayes[i])
```

```python
# Note, "gaussian" is the default argument for family. Added to be explicit. 
gauss_model = bmb.Model("en_ns ~ epsilon_critical", df_bayes, family="gaussian")
gauss_fitted = gauss_model.fit(draws=2000,chains=10, idata_kwargs={"log_likelihood": True})
gauss_model.predict(gauss_fitted, kind="response")
```

```python
az.summary(gauss_fitted)
```

```python
az.plot_trace(gauss_fitted)
```

```python
bmb.interpret.plot_predictions(gauss_model, gauss_fitted, 'epsilon_critical')
```

```python
az.plot_ppc(gauss_fitted, num_pp_samples=1000);
```

```python
t_model = bmb.Model("en_ns ~ epsilon_critical", df_bayes, family="t")
t_fitted = t_model.fit(draws=2000,chains=50, idata_kwargs={"log_likelihood": True})
t_model.predict(t_fitted, kind="response")
```

```python
az.summary(t_fitted)
```

```python
az.plot_trace(t_fitted, var_names=['Intercept','epsilon_critical'])
```

```python
az.plot_ppc(t_fitted, num_pp_samples=1000);
```

```python
# Generate the Bambi plot and get the figure and axes
fig, ax = plt.subplots()  # Replace with the actual Bambi plot function call
bmb.interpret.plot_predictions(t_model, t_fitted, 'epsilon_critical', ax=ax)

# Add your scatter plot to the same axes
ax.scatter(df_bayes['epsilon_critical'], df_bayes['en_ns'], color='lightpink')

# Add labels and legend
ax.set_xlabel('epsilon_critical')
ax.set_ylabel('en_ns')
#ax.set_title('Scatter Plot with Model Predictions Overlay')
ax.legend()

# Show plot
#plt.savefig("./posterior_enns_vs_epsilon.png")
plt.show()
```
